We will try to implement a simple system of symbolic differentiation on Python. The main type in our system will be a class `Expr` – an expression. All successor of the class should implement:

- `__call__` method which calculates a value of an expression in a specified context: the context bounds names of variables in the expression with specific values.
- `d` (Leibniz's notation of derivative) method which takes a name of a variable wrt (from “with respect to”) and returns an expression for a derivative with respect to this variable.

````python
class Expr:
    def __call__(self, **context):
        pass

    def d(self, wrt):
        pass
````

````python
class Const(Expr):
    def __init__(self, value):
        self.value = value
    
    def __call__(self, **context):
        return self.value
    
    def d(self, wrt):
        return Const(0)
    
    def __str__(self):
        return str(self.value)

class Var(Expr):
    def __init__(self, name):
        self.name = name
    
    def __call__(self, **context):
        return context[self.name]
    
    def d(self, wrt):
        if isinstance(wrt, Var):
            var_to_check = wrt.name
        else:
            var_to_check = wrt
        
        if self.name == var_to_check:
            return Const(1)
        else:
            return Const(0)
    
    def __str__(self):
        return self.name
````

2. Implement classes for binary operations: `Sum`, `Product`, and `Fraction`. Binary operations by definition work with exactly two operands that’s why a constructor for all binary classes will be the same. It’s convinient to take it out into a separate base class:


````python
class BinOp(Expr):
    def __init__(self, expr1: Expr, expr2: Expr) -> None:
        self.expr1, self.expr2 = expr1, expr2
````

````python
class Sum(BinOp):
    def __call__(self, **context):
        return self.expr1(**context) + self.expr2(**context)
    
    def d(self, wrt):
        # d(u + v)/dx = du/dx + dv/dx
        left_derivative = self.expr1.d(wrt)
        right_derivative = self.expr2.d(wrt)
        return Sum(left_derivative, right_derivative)
    
    def __str__(self):
        return f'(+ {self.expr1} {self.expr2})'


class Product(BinOp):
    def __call__(self, **context):
        return self.expr1(**context) * self.expr2(**context)
    
    def d(self, wrt):
        # d(uv)/dx = u(dv/dx) + v(du/dx)
        first_part = Product(self.expr1, self.expr2.d(wrt))
        second_part = Product(self.expr1.d(wrt), self.expr2)
        return Sum(first_part, second_part)
    
    def __str__(self):
        return f'(* {self.expr1} {self.expr2})'


class Fraction(BinOp):
    def __call__(self, **context):
        return self.expr1(**context) / self.expr2(**context)
    
    def d(self, wrt):
        # d(u/v)/dx = [v(du/dx) - u(dv/dx)] / v²
        numerator_part1 = Product(self.expr2, self.expr1.d(wrt))
        numerator_part2 = Product(self.expr1, self.expr2.d(wrt))
        numerator = Sum(numerator_part1, Negative(numerator_part2))
        
        denominator = Product(self.expr2, self.expr2)
        
        return Fraction(numerator, denominator)
    
    def __str__(self):
        return f'(/ {self.expr1} {self.expr2})'

````

3. Implement a class `Power` for an operation of an exponentiation operation. For simplicity you can expect that a power – is an expression consisting only of constants.

````python
class Power(BinOp):
    def __call__(self, **context):
        base_value = self.expr1(**context)
        exponent_value = self.expr2(**context)
        return base_value ** exponent_value
    
    def d(self, wrt):
        # d(u^n)/dx = n * u^(n-1) * du/dx
        n_minus_1 = Sum(self.expr2, Negative(Const(1)))
        u_power_n_minus_1 = Power(self.expr1, n_minus_1)
        n_times_power = Product(self.expr2, u_power_n_minus_1)
        result = Product(n_times_power, self.expr1.d(wrt))
        return result
    
    def __str__(self):
        return f'(** {self.expr1} {self.expr2})'

````
